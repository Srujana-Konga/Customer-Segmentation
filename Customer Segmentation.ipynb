## Data Cleaning

```python
# Finding missing values
df.isna().sum()
```

```python
# Filling missing values in 'Income' column with average value(Since less number of missing values)
df['Income'].fillna(df['Income'].mean(), inplace=True)
df
```

```python
df.isna().sum()
```

```python
# replacing values in marital_status column as two words has same meaning
replace_marital_status_values = {'Together': 'Married', 'Alone': 'Single'}
df['Marital_Status'] = df['Marital_Status'].replace(replace_marital_status_values)
df = df[~df['Marital_Status'].isin(['Absurd', 'YOLO'])]
df
```

```python
df['Marital_Status'].value_counts()
```

```python
# replpacing values in the column 'Education'
df.loc[df['Education'] == '2n Cycle', 'Education'] = 'Master'
```

```python
# removing unnecessary or columns which has constant values
df = df.drop(columns=['Z_CostContact','Z_Revenue','Response'])
```

```python
# Creating a new column 'TotalAmountSpent'
df['TotalAmountSpent'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
# Rearranging the columns
order = ['ID','Year_Birth','Marital_Status','Income','Kidhome','Teenhome','Dt_Customer','Recency','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases','NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth','AcceptedCmp1','AcceptedCmp2','AcceptedCmp3',   'AcceptedCmp4','AcceptedCmp5','Complain','TotalAmountSpent']
df = df[order]
df
```

## Exploratory Data Analysis

### Univariate Analysis

```python
numerical_columns = ['Year_Birth', 'Income', 'Kidhome', 'Teenhome', 'Recency', 
                     'MntWines', 'MntFruits', 'MntMeatProducts', 'NumCatalogPurchases', 
                     'NumStorePurchases', 'NumWebVisitsMonth', 'TotalAmountSpent']

#Histograms of numerical columns
for col in numerical_columns:
    plt.figure(figsize=(10, 4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Count plots for categorical columns
categorical_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 
                       'AcceptedCmp4', 'AcceptedCmp5', 'Complain']

for col in categorical_columns:
    plt.figure(figsize=(10, 4))
    sns.countplot(x=col, data=df)
    plt.title(f'Count Plot of {col}')
    plt.show()
```

### Bivariate Analysis

```python
# Correlation heatmap for numerical columns
plt.figure(figsize=(12, 8))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Box plots to see distribution of numerical columns by categorical columns
for col in categorical_columns:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=col, y='TotalAmountSpent', data=df)
    plt.title(f'Box Plot of TotalAmountSpent by {col}')
    plt.show()
```

```python
# Group by Marital_Status and calculate sum of TotalAmountSpent
amount_spent_by_marital = df.groupby('Marital_Status')['TotalAmountSpent'].sum().reset_index()

# Plotting the results
plt.figure(figsize=(10, 6))
sns.barplot(x='Marital_Status', y='TotalAmountSpent', data=amount_spent_by_marital, palette='viridis')
plt.title('Total Amount Spent by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
# Group by categories and calculate sum of Total Amount Spent
total_spent = df[['MntWines','MntFruits', 'MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']].sum()

# plot the graph
plt.figure(figsize=(12,8))
sns.barplot(x = total_spent.index, y= total_spent.values, palette = 'Pastel1')
plt.title("Total amount spent by product categories")
plt.xlabel("Product Categories")
plt.ylabel("Total amount spent")
plt.show()
```
